```python
from pathlib import Path
record_file = Path("./record_2.txt")
data = record_file.read_text(encoding="utf-8")
```

```python
time = []
position = []
analog_value = []

for line in data.strip().split("\n"):
    t, pos, digita, _ = map(float, line.split(" : "))
    if  t > 20:
        time.append(t)
        position.append(pos)
        analog_value.append(digita)

# Создаем фигуру и оси
fig, ax1 = plt.subplots()

# График положения
ax1.set_xlabel("Время (сек)")
ax1.set_ylabel("Положение", color="tab:blue")
ax1.plot(time, position, label="Положение", color="tab:blue")
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Вторая ось для аналогового значения
ax2 = ax1.twinx()
ax2.set_ylabel("Аналоговое значение", color="tab:red")
ax2.plot(time, analog_value, label="Аналоговое значение", color="tab:red", linestyle="dashed")
ax2.tick_params(axis="y", labelcolor="tab:red")

# Показываем график
fig.tight_layout()
plt.title("Положение и аналоговое значение от времени")
plt.show()
```
