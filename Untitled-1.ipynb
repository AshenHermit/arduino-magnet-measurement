```python
from pathlib import Path
record_file = Path("./record_2.txt")
data = record_file.read_text(encoding="utf-8")
```

```python
time = []
position = []
analog_value = []

for line in data.strip().split("\n"):
    t, pos, digita, _ = map(float, line.split(" : "))
    if  t > 20:
        time.append(t)
        position.append(pos)
        analog_value.append(digita)

#    
fig, ax1 = plt.subplots()

#  
ax1.set_xlabel(" ()")
ax1.set_ylabel("", color="tab:blue")
ax1.plot(time, position, label="", color="tab:blue")
ax1.tick_params(axis="y", labelcolor="tab:blue")

#     
ax2 = ax1.twinx()
ax2.set_ylabel(" ", color="tab:red")
ax2.plot(time, analog_value, label=" ", color="tab:red", linestyle="dashed")
ax2.tick_params(axis="y", labelcolor="tab:red")

#  
fig.tight_layout()
plt.title("     ")
plt.show()
```
